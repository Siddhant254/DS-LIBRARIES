```python
# From a pandas dataframe
batsman = pd.read_csv('/content/sharma-kohli.csv')
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# Multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# Labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
# Colors
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
# Linestyle
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
# Linewidth
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=4)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
# Markers
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',marker='o',markersize=8)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',marker='d')
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
# Legend
plt.plot(batsman['index'],batsman['V Kohli'],color='red',label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',label='Rohit Sharma')
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
plt.legend()
```

```python
# Limiting axes
plt.plot(batsman['index'],batsman['V Kohli'],color='red',label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',label='Rohit Sharma')
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.xlim(2012,2017)
plt.ylim(300,1000)
```

```python
# Grid
plt.plot(batsman['index'],batsman['V Kohli'],color='red',label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',label='Rohit Sharma')
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.xlim(2012,2017)
plt.ylim(300,1000)
plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='red',label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',label='Rohit Sharma')
plt.title("Rohit Sharma Vs Virat Kohli Runs comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.xlim(2012,2017)
plt.ylim(300,1000)
plt.grid()
plt.show()
```

## Annotations

```python
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['strike_rate'])
plt.xlabel("Average")
plt.ylabel("Strike-Rate")
plt.axhline(130,color='green')
plt.axvline(30,color='green')
for i in range(df.shape[0]):
  plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i],fontdict={'color':'black'})
```

## SUBPLOTS

```python
fig,ax = plt.subplots(figsize=(12,12),nrows=2,ncols=1)
ax[0].scatter(df['avg'],df['strike_rate'],c='green')
ax[1].scatter(df['avg'],df['runs'])

ax[0].set_title('Avg vs SR')
ax[0].set_xlabel("Avg")
ax[0].set_ylabel("strike_rate")

ax[1].set_title('Avg vs Runs')
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("Runs")
plt.show()
```

## Add_subplot

```python
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(df['avg'],df['strike_rate'])

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(df['avg'],df['runs'],color='yellow')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(df['avg'],color='green')

ax4 = fig.add_subplot(2,2,4)
ax4.hist(df['runs'],color='orange')
plt.show()
```

## SURFACE PLOTS

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z)
```

## Contour plots

```python
fig = plt.figure()
ax = plt.subplot()
p = ax.contourf(xx,yy,z)
fig.colorbar(p)
```
