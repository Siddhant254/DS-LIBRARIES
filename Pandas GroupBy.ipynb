```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("C:\\Users\\siddh\\OneDrive\\Desktop\\Data science data\\Pandas\\imdb-top-1000.csv")
movies.head(2)
```

```python
# Groupby
movies.groupby('Genre')
```

```python
# Top 3 genres by total earning
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
# Genre with highest avg IMDB rating
movies.groupby('Genre').mean()['IMDB_Rating'].sort_values(ascending=False).head(1)
```

```python
# Director with most popularity
movies.groupby('Director').sum()['No_of_Votes'].sort_values(ascending=False).head(1)
```

```python
# Highest Rated movie of each genre
movies.groupby('Genre').max()[['Series_Title','IMDB_Rating']]
```

```python
# Number of Movies done by each actor
movies.groupby('Star1').count()['Series_Title'].sort_values(ascending=False)
```

## Groupby attributes

```python
# total groups
len(movies.groupby('Genre'))
```

```python
# Items in each group
movies.groupby('Genre').size()
```

```python
# First_item
movies.groupby('Genre').first()
```

```python
# last item
movies.groupby('Genre').last()
```

```python
# Horror movies
movies.groupby('Genre').get_group('Horror')
```

```python
# AGG METHOD
movies.groupby('Genre').agg({'Runtime':'mean',
                             'IMDB_Rating':'mean',
                            'No_of_Votes':'sum',
                            'Gross':'sum'})
```

```python
# Number of Movies starting with A in each genre

def start_a(group):
    return group['Series_Title'].str.startswith('A').sum()

movies.groupby('Genre').apply(start_a)
```

### MULTIPLE COLUMNS

```python
duo = movies.groupby(['Director','Star1'])
duo
```

```python
# Most earning actor director combo
duo.sum()['Gross'].sort_values(ascending=False).head(1).reset_index()
```

```python
# Aggregate
duo.agg(['mean']).head()
```

## DELIVERIES DATASET

```python
ipl = pd.read_csv("C:\\Users\\siddh\\OneDrive\\Desktop\\Data science data\\Pandas\\deliveries.csv")
ipl.head()
```

```python
# TOP 10 BATSMAN IN TERMS OF RUNS SCORED
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).reset_index().head(10)
```

```python
# Batsman with max no. of sixes
six = ipl[ipl['batsman_runs']==6]
six.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).reset_index().head(5)
```

```python
# Virat kohli's record against all teams
kohli = ipl[ipl['batsman']=='V Kohli']
kohli.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```
